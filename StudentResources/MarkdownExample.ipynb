##Tables

| Col 1        | Col 2       | Col3         |
| ----------   | :--------:  | -----------: |
| this         | this        | this         |
| column       | column      | column       |
| is           | is          | is           |
| left-aligned | centered    | right-aligned|

            OR

Col 1 | Col 2 | Col3
--- | :---: | ---:
this | this | this
column | column | column
is | is | is
left-aligned | centered | right-aligned

##Lists

1. ordered list item 1
2. ordered list item 2
4. ordered list item 4 (Notice that the numbers in the list don't have to be in order.)
* ordered list item * (and that it doesnt matter what symbol)


* Unordered  lists can use asterisks
- or minuses/dashes
+ or pluses
