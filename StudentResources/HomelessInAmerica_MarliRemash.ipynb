## Homeless In America

Inspiration from: https://www.facinghomelessness.org/

DataSet from: https://jupyter.grinnell.edu/user/remashma/notebooks/HomelessInAmerica_MarliRemash.ipynb

```python
%matplotlib inline

import pandas as pd
import statistics as st
import matplotlib
import numpy as np

matplotlib.style.use('ggplot')

filename_2016 = "https://raw.githubusercontent.com/mashmouse/DataScienceForSocialGoodJupyterNotebooks/master/Data/Homeless-by-State_2016_fixed.csv"

data_2016 = pd.read_csv(filename_2016)
data_2015 = pd.read_csv(filename_2015)

df16 = pd.DataFrame(data_2016)

total_pop = df16['Total Homeless, 2016'][54]

df16_ind = pd.DataFrame(data_2016)
df16_ind.index = df16_ind['State']
#df16_ind[:54].plot.pie(y='Total Homeless, 2016', figsize=(15, 15))

df16_percent = df16_ind
df16_percent['State Percent of Total'] = round(df16_percent['Total Homeless, 2016'] / total_pop * 100, 2)
df16_percent['State Percent and Name'] = df16_percent['State'] + ' ' + df16_percent['State Percent of Total'].apply(str)
df16_percent.index = df16_percent['State Percent and Name']
df16_percent[:54].plot.pie(y='Total Homeless, 2016', figsize=(15, 15))


df16[:54].plot.bar(x='State', y=['Total Homeless, 2016', 'Number of CoCs'], figsize=(20, 15), subplots=True)


```
