```python
#This is a comment.

#This allows us to display graphs in our notebook
%matplotlib inline

#Import these 
import pandas as pd
import numpy as np
import statistics as st
import matplotlib

matplotlib.style.use('ggplot')

#Reading Data from Files
filename = "PoweshiekTemperaturePredictionsF.csv"
data = pd.read_csv(filename)
df = pd.DataFrame(data)
```

## Line Graph/Plot

Here is the structure of writting code for a line plot:

DataFrame.plot(x=*x-coordinates*, y=*y-coordinates*)

```python
df.plot(x='year', y='rcp45median')
```

## Two Plots in One Frame

In order to display multiple line plots on the same graph:

DataFrame.plot(x=*x-coordinates*, y=*[y-coords-col-1, y-coords-col-2*, ... y-coords-col-n])

```python
df.plot(x='year', y=['rcp45median', 'rcp45min', 'rcp45max'])
```

## Plots in Nearby Frames

In order to display multiple graphs in separate frames:


DataFrame.plot(x=*x-coordinates*, y=*[y-coords-col-1, y-coords-col-2*, ... y-coords-col-n], subplots=True)

```python
df.plot(x='year', y=['rcp45min', 'rcp45max', 'rcp45median'], subplots=True)
```

## Making Pie Charts

Sometimes choosing a different type of visualizaton helps us understand the data better... this is not one of those times (there are 94 different data points in the given series).

In order to display a pie chart using your data:

DataFrame-column.plot.pie()  OR  Series.plot.pie()

```python
df['rcp45min'].plot.pie()
```

## Bar Graphs

Lets try another type of visualization... it still might be a bit incomprehensible. 

In order to display a bar graph using your data:

DataFrame-column.plot.pie()  OR  Series.plot.pie()

```python
df['rcp45min'].plot.bar()
```

Not that it is very necessary here... but if desired there can be multple Series in one bar graph. 

```python
df.plot.bar()
```

## Scatter Plots

Scatter Plots are a bit odd. Rather than having a Series.plot.*graph-type* format, they are structured this way:

DataFrame.plot.scatter(x=*x-coordinates*, y=*y-coordinates*, *otional-stuff...*)  

Prominent Optional Stuff:
* c= color (number value) 
* s= size (radius of the points) 

```python
df.plot.scatter(x='year', y='rcp45min', c='rcp45max', s=df['rcp45median'])
```

## Hexagonal Bin Plot

Here is yet another way to display data:

DataFrame.plot.hexbin(x=*x-coord*, y=*y-coord*, gridsize=*int*)

NOTE: gridsize is the number of major hexagon shaped partitions made in the graph (hight number = more specificity)

```python
df.plot.hexbin(x='year', y='rcp45min', gridsize=20)
```

## Box Plot

Here is another way to display max, min, and median:

DataFrame.plot.box(y=[DataFrame.Series, DataFrame.Series... etc.])

```python
df.plot.box(y=['rcp45min', 'rcp45max', 'rcp45median'])
```

## Area Plots

Here is some syntax:

DataFrame.plot.area()

```python
df.plot.area()
```

Here is a way that changes the alpha value of the graphs...

```python
df.plot.area(stacked=False)
```
