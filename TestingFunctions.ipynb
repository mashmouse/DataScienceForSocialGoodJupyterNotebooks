```python
def multiply_lst (lst, num):
    index = 0
    for item in lst:
        lst[index] = num * item
        index = index + 1
    return lst
print(multiply_lst(["hello", "world"], 2))
```

```python
def sum_lst(lst1, lst2):
    if len(lst1) != len(lst2):
        print("Error: the lists have different lengths")
    else:
        index = 0
        for item in lst1:
            lst1[index] = lst1[index] + lst2[index]
            index = index + 1
        return lst1

sum_lst([1,2,3,4], [1,2,3,4,5])
```

```python
from statistics import median
    
def interquartile(lst):
    sorted(lst)
    index = median(lst)
    first_lst = lst[:,index]
    second_lst = lst[index, :]
    first_quartile = median(first_lst)
    second_quartile = median(second_lst)
    return (first_quartile, second_quartile)

print(interquartile([1,4,56,7,8,3,56,7,3,3,4,7,8]))
```

```python
def greetings(time):
    if time < 1200:
        print("Good morning!")
    elif time < 1800:
        print("Good afternoon!")
    elif time < 2100:
        print("Good evening!")
    elif time < 2400: 
        print("Good night!")
    else:
        print("Silly goose! That's not a time!")
greetings(2300)
```
